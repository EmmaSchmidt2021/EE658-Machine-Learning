```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
digits=load_digits()
```

```python
print("Data size", digits.data.shape)
```

```python
print("Number of Labels", digits.target.shape)
```

```python
fig = plt.figure(figsize=(20,26))
fig.subplots_adjust(left=0, right=1, bottom=0, top=1, hspace=0.05, wspace=0.35)
for index,(image,label) in enumerate(zip(digits.data[0:20], digits.target[0:20])):
    plt.subplot(4,5,index+1)
    plt.imshow(np.reshape(image, (8,8)), cmap=plt.cm.gray)
    plt.title('Training: %i\n' %label, fontsize=20)
plt.show()
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(digits.data, digits.target, test_size=0.25, random_state=2)
```

```python
from sklearn.linear_model import LogisticRegression
logisticReg = LogisticRegression()
logisticReg.fit(x_train, y_train)
```

```python
predictions = logisticReg.predict(x_test)
score = logisticReg.score(x_test, y_test)
print(score)
```

```python
import matplotlib.pyplot as plt
from sklearn import metrics
cm=metrics.confusion_matrix(y_test, predictions)
```
