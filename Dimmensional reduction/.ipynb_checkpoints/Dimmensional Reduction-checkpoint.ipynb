```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
np.random.seed(27)
# Create the 2-D data set: x1, x2
m = 30
x1 = np.array(50 * np.random.rand(m))
x2 = 10 + 2 * x1 + 70 * np.random.rand(m)
```

```python
# Plot the data set
plt.plot(x1, x2, 'bo')
plt.axhline(linewidth=0.5, color='r')
plt.axvline(linewidth=0.5, color='r')
plt.xlabel("x1")
plt.ylabel("x2")
plt.show()
```

```python
#only represent x1 - project X1 onto the x axis - ignore x2(y coordinate)
plt.plot(x1, x2, 'bo')
plt.plot(x1, np.zeros([len(x1)]), 'ro')
plt.axhline(linewidth=0.5, color='r')
plt.axvline(linewidth=0.5, color='r')
plt.xlabel("x1")
plt.ylabel("x2")
plt.show()
#but is this a reasonable approximation? - do linear transform without effecting data
```

```python
# The normalized data centered at the origin -TRANSLATE ALL THE POINTS
# did not change relationship between points
x1_mean = x1.mean()
x2_mean = x2.mean()
print('x1 mean =', x1_mean)
print('x2 mean =', x2_mean)
x1n = x1 - x1_mean
x2n = x2 - x2_mean
```

```python
# Plotting the normalized data
plt.plot(x1n, x2n, 'bo')
ax = plt.gca()
ax.annotate("(0,0)", (0.1, -4))
plt.axhline(linewidth=0.5, color='r')
plt.axvline(linewidth=0.5, color='r')
```

```python
# Compute the Regression line
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(x1n.reshape(1,-1).T, x2n)
print('Regression Line: Intercept=', lin_reg.intercept_, 'Slope=', lin_reg.coef_)
```

```python
# Regression Line: Intercept= 5.1661e-15    Slope= [1.53467]
plt.plot(x1n, x2n, 'ro')
plt.plot([-25,30], [lin_reg.coef_[0]*(-50), 
lin_reg.coef_[0]*(50)],  'b-' )
plt.axhline(linewidth=0.5, color='green')
plt.axvline(linewidth=0.5, color='green')
plt.show()
```

```python
# Computing the SSD [Sum of Squared Differences]
def ssd(x, y):
    m = len(x)
    y_mean = y.mean()
    x_mean = x.mean()
    g1 = np.square(y).sum() - (m * y_mean ** 2)
    g2 = np.square(x).sum() - (m * x_mean ** 2)
    g3 = m * x_mean * y_mean - (x* y).sum()
    B = 0.5 * (g1-g2) / g3
    b1 = -B + np.sqrt(B**2 + 1)
    b2 = -B - np.sqrt(B**2 + 1)
    intercept1 = y_mean - b1 * x_mean
    intercept2 = y_mean - b2 * x_mean
    return intercept1, intercept2, b1, b2
```

```python
intercept1, intercept2, slope1, slope2 = ssd(x1n, x2n)
print('Slope1:',  round(slope1,2), 'intercept1:', round(intercept1,2))
print('Slope2:',  round(slope2,2), 'intercept2:', round(intercept2,2))
```

```python
# Plot the SSD and the Regression lines
plt.plot(x1n, x2n, 'ro')
plt.plot([-25,30], [lin_reg.coef_[0]*(-25), lin_reg.coef_[0]*(30)], 'b-', label='Linear Regression')
plt.plot([-25,30], [slope1*(-25), slope1*30], 'g-', label='SSD-1')
plt.plot([-25,30], [slope2*(-25), slope2*30], 'g-', label='SSD-2')
ax = plt.gca()
ax.set_aspect('equal')
ax.legend()
plt.axhline(linewidth=0.5, color='green')
plt.axvline(linewidth=0.5, color='green')
plt.show()
```

```python
# Unit vector in the direction of the SSD line (PC1)
a1 = 1./np.sqrt(1+ slope1**2)
b1 = a1 * slope1
unit_vector_1 = [a1, b1]
print("PC1:", round(a1, 2), round(b1, 2))
print(a1**2 + b1**2)
```

```python
# Unit vector in the direction of the line perpendicular to SSD (PC2)
a2 = 1./np.sqrt(1+ slope2**2)
b2 = a2 * slope2
unit_vector_2 = [a2, b2]
print("PC2:", round(a2, 2), round(b2, 2))
print(a2**2 + b2**2)
```

```python
# Using the Python library, we can find the Eigenvalues and Eigenvectors
X = np.vstack((x1n, x2n))
sigma = np.dot(X, X.T)    # the covariance matrix
v, W = np.linalg.eig(sigma)
print("v=", v); print("W=", W)
```
